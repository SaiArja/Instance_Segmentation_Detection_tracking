***Checking the image and mask***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
def plot_img_mask(img, mask):
  img = cv2.imread(img)
  mask = cv2.imread(mask)

  fig = plt.figure(figsize=(18, 8))
  rows = 1
  columns = 2

  fig.add_subplot(rows, columns, 1)
  plt.imshow(img)
  plt.axis('on')
  plt.title("Image")
    
  # Adds a subplot at the 2nd position
  fig.add_subplot(rows, columns, 2)
  plt.imshow(mask)
  plt.axis('on')
  plt.title("Mask")
```

```python
import os
import pandas as pd
def get_dataframe(source_dir, name):
  '''
  This Function will take two parameters the source_dir and name and return all the image paths 
  of the images in a dataframe with name as coloumn name of the Dataframe
  '''
  source_dir = source_dir
  #source_dir is the directrory where images are saved
  ann_paths=[]
  filelist=os.listdir(source_dir)
  for f in filelist:
      fpath=os.path.join(source_dir,f)
      #fpath will take us to the subfolders in the source_dir folder
      ann_dir =os.listdir(fpath)
      for ann in ann_dir:
        #This look will loop around the images of the images in the subfolder
        annpath=os.path.join(fpath,ann)
        ann_paths.append(annpath)
    

  Fseries=pd.Series(ann_paths, name=name)
  #creating a Series from the list
  dataframe=pd.concat([Fseries], axis=1)
  #creating a dataframe
  return dataframe
```

```python
df1 = get_dataframe('/content/train/Annotations', 'Annotations')
df2 = get_dataframe('/content/train/JPEGImages', 'Images')
```

```python
df1 = df1.sort_values("Annotations").reset_index().drop('index', axis = 1)
#the values are sorted and the index is droped as it will be symmetrical with the images in df2
```

```python
df1.head(10)
```

```python
# df2.head()
df2 = df2.sort_values("Images").reset_index().drop('index', axis = 1)
#the values are sorted and the index is droped as it will be symmetrical with the Annotations in df1
```

```python
df2.head(10)
```

```python
df_train = pd.concat([df1, df2],axis = 1)
#making the annotations dataframe and Inmages dataframe in to one
```

```python
df_train = df_train.drop('level_0', axis = 1)
#dropping index
```

***Plotting a Random Image***

```python
import random
def plot():
  n = random.randint(0,len(df_train))
  image = df_train.iloc[df_train['level_0'][n]].Images          
  mask = df_train.iloc[df_train['level_0'][n]].Annotations
  plot_img_mask(img = image, mask = mask)
```

```python
plot()
```
